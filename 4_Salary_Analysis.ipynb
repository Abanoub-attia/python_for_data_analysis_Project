# Salary Analysis

## Prepare the Data 

### Importing Libraries

```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
```

### Loading Data

```python
dataset= load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas() 
```

### Data Cleanup

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

## Salary Distribution by Job Title

### Filter for US

```python
df_US = df[df['job_country'] == 'United States'].dropna(subset=['salary_year_avg']).copy()
```

```python
df_US
```

### Select Top 6 Roles

```python
job_titles = df_US['job_title_short'].value_counts().index[:6].tolist()
```

```python
job_titles
```

```python
df_US_top6 = df_US[df_US['job_title_short'].isin(job_titles)]
```

```python
df_US_top6
```

### Group by Jop Title

```python
job_order = df_US_top6.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
```

```python
job_order
```

## Plot Salary Distributions

```python
sns.boxplot(data=df_US_top6, x='salary_year_avg', y='job_title_short', order=job_order)
sns.set_theme(style='ticks')

# this is all the same
plt.title('Salary Distributions in the United States')
plt.xlabel('Yearly Salary (USD)')
plt.ylabel('')
plt.xlim(0, 600000) 
ticks_x = plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K')
plt.gca().xaxis.set_major_formatter(ticks_x)
plt.show()
```

## Investigate Median Salary Vs Skill for Data Analysts

### Filter for US Data Analyst Role

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()

df_DA_US = df_DA_US.dropna(subset=['salary_year_avg'])

df_DA_US = df_DA_US.explode('job_skills')
```

```python
df_DA_US[['salary_year_avg', 'job_skills']]
```

### Get the Most Demanded Skills

```python
df_DA_top_pay = df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='median', ascending=False)

df_DA_top_pay = df_DA_top_pay.head(10)
```

```python
df_DA_top_pay
```

### Get the Highest Paid Skills

```python
df_DA_skills = df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False)

df_DA_skills = df_DA_skills.head(10).sort_values(by='median', ascending=False)
```

```python
df_DA_skills
```

## Determine The Highest Paid Skills and Most Demanded Skills

```python
fig, ax = plt.subplots(2,1)

sns.set_theme(style='ticks')

#df_DA_top_pay[::-1].plot(kind='barh', y='median', ax=ax[0], legend=False) #[::-1] = ax[0].invert_yaxis()
sns.barplot(data=df_DA_top_pay, x='median', y=df_DA_top_pay.index, ax=ax[0], hue='median', palette='dark:b_r')
ax[0].set_title('Top 10 Highest Paid Skills for Data Analysis')
ax[0].set_ylabel('')
ax[0].set_xlabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
ax[0].legend().remove()

#df_DA_skills[::-1].plot(kind='barh', y='median', ax=ax[1], legend=False)
sns.barplot(data=df_DA_skills, x='median', y=df_DA_skills.index, ax=ax[1], hue='median', palette='light:b')
ax[1].set_title('Top 10 Most In-Demand Skills for Data Analysis')
ax[1].set_ylabel('')
ax[1].set_xlabel('')
ax[1].set_xlim(ax[0].get_xlim())
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
ax[1].legend().remove()

fig.tight_layout()
fig.show()
```
